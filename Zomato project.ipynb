Zomato Data Analysis Porject


Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2 - Create the data frame

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

Convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

Type of Restaurant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Count", c="red", size=20)
```

Conclusion - Majority of the Restaurant falls in dining category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

Conclusion - Dining Restaurants has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins =5)
plt.title("ratings distribution")
plt.show()
```

Conclusion - The majority Restaurants recieved ratings from 3.5 to 4

Average Order Spending by Couples

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Conclusion - The majority of couples prefer restaurants with an approximate cost of 300 rupees

Which mode receives maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

Conclusion - Offline order received lower rating in comparision to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

Conclusion - Dining Restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients preffered to place orders in person at restaurants, but prefer online ordering at cafes.
